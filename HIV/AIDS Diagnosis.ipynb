**Boxplot** is used to check the outliers of the dataset

```python
plt.figure(figsize=(5,5))
sns.boxplot(df3,showmeans=True)
```

 Dataset contain outliers.To remove outlier values from a DataFrame in Python, we can use various statistical methods. One common approach is to identify outliers based on the **Z-score or IQR** (Interquartile Range).  Here, I'll show you how to remove outliers based on the Z-score using the scipy library.

```python
from scipy import stats

# Define a z-score threshold (e.g., 3.0) to identify outliers
z_score_threshold = 3.0

# Calculate the absolute z-scores for all columns
z_scores = np.abs(stats.zscore(df3))

# Create a boolean mask where True represents an outlier in any column
outlier_mask = (z_scores > z_score_threshold).any(axis=1)

# Remove rows with outliers
df4 = df3[~outlier_mask]
```

### **x,y saperation**

```python
x=df4.drop(['TOTAL NUMBER OF AIDS DIAGNOSES'],axis=1).values
x
```

```python
y=df4['TOTAL NUMBER OF AIDS DIAGNOSES'].values
y
```

### **Train Test split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

### **Model Creation:**
###    2 different regression algorithms are used for the model creation.
###       
###        1. Decision tree
###       2.Random forest

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
dt=DecisionTreeRegressor()
rf=RandomForestRegressor(n_estimators=100)
lst=[dt,rf]
```

### **Performance measurment**

```python
from sklearn.metrics import r2_score,mean_absolute_percentage_error
for i in lst:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  y_pred
  print('r2_score=',r2_score(y_test,y_pred))
  print('MAPE =',mean_absolute_percentage_error(y_test,y_pred))
```

### **Conclusion**
Among the two model 'Random Forest regression' model is better suit for this HIV/AIDS diagnoses dataset,which provides better value for both r2_score and mean absolute percentage error.
